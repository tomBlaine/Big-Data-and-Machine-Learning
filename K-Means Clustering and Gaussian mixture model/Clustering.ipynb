*Task 2.1*

```python
import numpy
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
data1 = numpy.load("Iris_data.npy")
```

```python
numpy.shape(data1)
```

```python
data_labels = numpy.load("Iris_labels.npy")
```

```python
print(data_labels)
```

Iris data contains:\
150 samples\
4 features

Iris labels contains:\
150 values, Integer between 0 and 2\
It denotes the true species membership of samples

```python
plt.figure()
plt.scatter(data1[:,0], data1[:,1], c=data_labels)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Sepal length compared to width')
plt.show() 
```

*Task 2.2*

```python
kmeans = KMeans(n_clusters=3).fit(data1)
```

this initialises a kmeans object with n=3 clusters and passes the iris data set

```python
p_labels = kmeans.predict(data1)
```

this predicts the cluster membership of each data sample 

```python
kmeans.cluster_centers_
```

```python
plt.figure()
plt.scatter(data1[:,0], data1[:,1], c=p_labels)
plt.plot(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 'r+')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Sepal length compared to width')
plt.show() 
```

this plots a scatter graph for iris data set and colours samples their predicted cluster membership

```python
kmeans2 = KMeans(n_clusters=3,n_init=3000, init = 'random').fit(data1)
```

this initialises a new kmeans object with n=3 clusters, runs the kmeans algorithm 3000 times for each centroid and begins with a random placement of centroids

```python
p1_labels = kmeans2.predict(data1)
```

```python
plt.figure()
plt.scatter(data1[:,0], data1[:,1], c=p1_labels)
plt.plot(kmeans2.cluster_centers_[:,0], kmeans2.cluster_centers_[:,1], 'r+')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Sepal length compared to width')
plt.show() 
```

this plots a scatter graph for iris data set and colours samples their predicted cluster membership (centroids are marked in red)

*Task 2.3*

```python
gm = GaussianMixture(n_components=3, n_init=10).fit(data1)
```

```python
gm_labels=gm.predict(data1)
```

```python
plt.figure()
plt.scatter(data1[:,0], data1[:,1], c=gm_labels)
plt.plot(gm.means_[:,0], gm.means_[:,1], 'r+')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Sepal length compared to width')
plt.show() 
```

```python
predict_prob_data = gm.predict_proba(data1)
```

```python
plt.figure()
plt.scatter(data1[:,0], data1[:,1], cmap='viridis', c=predict_prob_data[:,0])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Component 0')
plt.colorbar()
plt.show() 
```

```python
plt.figure()
plt.scatter(data1[:,0], data1[:,1], cmap='viridis', c=predict_prob_data[:,1])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Component 1')
plt.colorbar()
plt.show() 
```

```python
plt.figure()
plt.scatter(data1[:,0], data1[:,1], cmap='viridis', c=predict_prob_data[:,2])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Component 2')
plt.colorbar()
plt.show() 
```

n_init = The number of initializations to perform where the best results are kept.
